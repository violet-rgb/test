```python
#load the libraries 
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
```

```python
# Data Understanding - Step-by-step checks and comments
# 1) Display basic dataset info: shape, columns and data types
# 2) Preview data: head and tail
# 3) Generate numeric and full statistical summaries
# 4) Identify numerical vs categorical columns and show example value counts

print('--- Data Understanding ---')

# Shape
print('\nDataset shape:')
print(df.shape)

# Columns and dtypes
print('\nColumns:')
print(list(df.columns))

print('\nData types:')
print(df.dtypes)

# Head and tail
print('\nFirst 5 rows (head):')
display(df.head())

print('\nLast 5 rows (tail):')
display(df.tail())

# Statistical summaries
print('\nStatistical summary (numeric columns):')
display(df.describe())

print('\nStatistical summary (all columns):')
display(df.describe(include="all"))

# Identify numeric and categorical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
print('\nNumeric columns:')
print(numeric_cols)
print('\nCategorical (non-numeric) columns:')
print(categorical_cols)

# Show top value counts for categorical columns, if present
if categorical_cols:
    print('\nTop value counts for categorical columns:')
    for col in categorical_cols:
        print(f"\nColumn: {col}")
        display(df[col].value_counts().nlargest(10))
else:
    print('\nNo non-numeric categorical columns detected.')
```

```python
# Missing Value Analysis and Imputation
# 1) Check missing counts and percentage
# 2) Explain why imputation is necessary
# 3) Compute chosen imputation values (median for numeric, mode for categorical)
# 4) Apply imputation if missing values exist, validate, and save

print('--- Missing Value Analysis ---')

missing_counts = df.isnull().sum()
missing_pct = (df.isnull().mean() * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing_counts, 'missing_pct': missing_pct})
display(missing_df)

# Explanation
print('\nWhy imputation is necessary:')
print('- Preserves dataset size (avoids dropping many rows)')
print('- Prevents model training errors on algorithms that cannot handle NaNs')
print('- Reduces bias introduced by arbitrary row removal; choosing a robust imputation reduces distortion')

# Identify numeric and categorical columns (recompute to be safe)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

# Choose imputation methods and justify
print('\nChosen imputation:')
print('- Numeric columns: median (robust to outliers)')
print('- Categorical columns: mode (most frequent category)')

# Show the chosen imputation values (but do not apply yet)
medians = df[numeric_cols].median()
print('\nMedians for numeric columns:')
display(medians.to_frame('median'))

modes = {col: (df[col].mode().iloc[0] if not df[col].mode().empty else None) for col in categorical_cols}
if categorical_cols:
    print('\nModes for categorical columns:')
    print(modes)

# Validation step: apply only if there are missing values
if int(missing_counts.sum()) == 0:
    print('\nNo missing values detected — no imputation applied.')
else:
    print(f"\nApplying imputation to {int(missing_counts.sum())} missing values...")
    df_imputed = df.copy()
    for col in numeric_cols:
        df_imputed[col] = df_imputed[col].fillna(medians[col])
    for col in categorical_cols:
        if df_imputed[col].isnull().any():
            df_imputed[col] = df_imputed[col].fillna(modes[col])

    print('\nPost-imputation missing counts:')
    display(df_imputed.isnull().sum())

    # Save imputed file for downstream steps
    df_imputed.to_csv('heart_imputed.csv', index=False)
    print('\nSaved imputed dataset: heart_imputed.csv')
```

```python
# Duplicate Analysis
# 1) Detect duplicate rows and show a sample
# 2) Investigate duplicates to ensure they are true duplicates
# 3) Remove duplicates if necessary and show counts before/after

print('--- Duplicate Analysis ---')

# Detect duplicates (exact row duplicates)
duplicate_mask = df.duplicated(keep=False)
dup_rows = df[duplicate_mask]
print('\nTotal duplicate rows (including all copies):', len(dup_rows))

if len(dup_rows) > 0:
    print('\nShowing up to 20 duplicated rows:')
    display(dup_rows.head(20))

    # Investigate whether duplicates differ on important columns
    print('\nChecking duplicate groups summary (by duplicated subset):')
    dup_groups = dup_rows.groupby(list(df.columns)).size().reset_index(name='count')
    display(dup_groups.head(20))

    # Remove exact duplicates (keep first occurrence)
    before = df.shape[0]
    df_no_dup = df.drop_duplicates(keep='first').reset_index(drop=True)
    after = df_no_dup.shape[0]
    print(f'\nRows before deduplication: {before}, after: {after}, removed: {before - after}')

    # Replace df with deduped version for downstream steps
    df = df_no_dup
else:
    print('\nNo exact duplicate rows detected.')

# Save deduped file
df.to_csv('heart_deduped.csv', index=False)
print('\nSaved deduplicated dataset: heart_deduped.csv')
```

```python
# Target Variable Analysis
# 1) Check class distribution and percentages
# 2) Plot countplot (requires seaborn)
# 3) Detect imbalance and demonstrate three remedies: SMOTE, RandomOverSampler, RandomUnderSampler
# 4) Explain recommended approach and save resampled datasets

import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

print('--- Target Variable Analysis ---')

target_col = 'target'

# Class distribution
counts = df[target_col].value_counts()
counts_pct = (df[target_col].value_counts(normalize=True) * 100).round(2)
print('\nClass counts:')
print(counts)
print('\nClass percentages:')
print(counts_pct)

# Plot countplot
plt.figure(figsize=(5,4))
sns.countplot(x=target_col, data=df)
plt.title('Target class distribution')
plt.show()

# Detect imbalance
minority = counts.min()
majority = counts.max()
ratio = minority / majority
print(f'\nMinority/Majority ratio: {ratio:.3f}')
imbalanced = ratio < 0.5
print('Imbalanced:' , imbalanced)

# If imbalanced, demonstrate SMOTE / oversampling / undersampling
if imbalanced:
    print('\nApplying resampling techniques for demonstration (SMOTE, RandomOverSampler, RandomUnderSampler)')
    X = df.drop(columns=[target_col])
    y = df[target_col]

    # Keep only numeric features for resampling (SMOTE requires numeric inputs)
    X_num = X.select_dtypes(include=["number"]).copy()

    try:
        from imblearn.over_sampling import SMOTE
        from imblearn.over_sampling import RandomOverSampler
        from imblearn.under_sampling import RandomUnderSampler
        has_imblearn = True
    except Exception:
        has_imblearn = False

    if not has_imblearn:
        print('\nimblearn not installed — cannot run SMOTE/oversampling/undersampling.\nInstall with: pip install imbalanced-learn')
    else:
        # SMOTE
        sm = SMOTE(random_state=42)
        X_sm, y_sm = sm.fit_resample(X_num, y)
        print('\nSMOTE result class counts:')
        print(Counter(y_sm))
        df_smote = pd.concat([pd.DataFrame(X_sm, columns=X_num.columns), pd.Series(y_sm, name=target_col)], axis=1)
        df_smote.to_csv('heart_smote.csv', index=False)

        # Random Oversampling
        ros = RandomOverSampler(random_state=42)
        X_over, y_over = ros.fit_resample(X_num, y)
        print('\nRandomOverSampler result class counts:')
        print(Counter(y_over))
        df_over = pd.concat([pd.DataFrame(X_over, columns=X_num.columns), pd.Series(y_over, name=target_col)], axis=1)
        df_over.to_csv('heart_oversampled.csv', index=False)

        # Random Undersampling
        rus = RandomUnderSampler(random_state=42)
        X_under, y_under = rus.fit_resample(X_num, y)
        print('\nRandomUnderSampler result class counts:')
        print(Counter(y_under))
        df_under = pd.concat([pd.DataFrame(X_under, columns=X_num.columns), pd.Series(y_under, name=target_col)], axis=1)
        df_under.to_csv('heart_undersampled.csv', index=False)

        # Explanation and recommendation
        print('\nRecommendation:')
        print('- SMOTE: creates synthetic minority samples; good when minority class is small and features are numeric/continuous.')
        print('- Oversampling (Random): duplicates minority samples; simple but may overfit.')
        print('- Undersampling (Random): reduces majority class; may lose information and is risky with small datasets.')
        print('\nChosen approach for most cases: SMOTE (when numeric features present and dataset moderately imbalanced) — it balances classes without dropping majority examples.')
else:
    print('\nClasses are reasonably balanced; resampling not required.')

```

```python
# 5–14: Analysis, Encoding, Scaling, Feature Selection, Modeling, Evaluation
# This cell performs univariate analysis, outlier checks, bivariate analysis,
# encoding, scaling (where needed), feature selection, train-test split,
# trains multiple classifiers, and reports evaluation metrics and comparisons.

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC

# Optional XGBoost
try:
    from xgboost import XGBClassifier
    has_xgboost = True
except Exception:
    has_xgboost = False

print('--- Univariate Analysis ---')
num_cols = df.select_dtypes(include=['number']).columns.tolist()
num_cols.remove('target') if 'target' in num_cols else None

for col in num_cols:
    plt.figure(figsize=(10,3))
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=False)
    plt.title(f'Histogram of {col}')
    plt.subplot(1,2,2)
    sns.kdeplot(df[col], shade=True)
    plt.title(f'KDE of {col} (skew={df[col].skew():.2f})')
    plt.show()
    print(f'{col} skewness: {df[col].skew():.3f}\n')

print('Skewness explanation:')
print('- Skewness ~0: symmetric; >0: right-skewed; <0: left-skewed. Strong skew may need transformation (log/sqrt).')

# 6. Outlier Detection
print('\n--- Outlier Detection (Boxplots & IQR) ---')
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

outliers = {}
for col in num_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    mask = (df[col] < lower) | (df[col] > upper)
    outliers[col] = mask.sum()
print('\nOutlier counts (IQR method):')
print(outliers)

print('\nOutlier handling:')
print('- Investigate per-feature; for this dataset we clipped earlier during preprocessing. Remove only if erroneous.')

# 7. Bivariate & Multivariate Analysis
print('\n--- Correlation heatmap ---')
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation matrix')
plt.show()

# Pairplot may be heavy; show for a small subset
subset = num_cols[:6] if len(num_cols) > 6 else num_cols
print('\nPairplot for a subset of numeric features:')
plt.figure()
sns.pairplot(df[subset + ['target']], hue='target', corner=True)
plt.show()

# Relationship with target: group means
print('\nFeature means by target:')
df.groupby('target')[num_cols].mean()

# Feature importance via RandomForest (simple)
X = df.drop(columns=['target'])
y = df['target']
X_num = X.select_dtypes(include=['number'])
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_num, y)
importances = pd.Series(rf.feature_importances_, index=X_num.columns).sort_values(ascending=False)
print('\nTop feature importances (RandomForest):')
print(importances.head(10))

# 8. Encoding
print('\n--- Encoding ---')
cat_cols = df.select_dtypes(exclude=['number']).columns.tolist()
print('Categorical columns:', cat_cols)

df_encoded = df.copy()
if cat_cols:
    # Use one-hot for nominal with few categories; label encode for ordinal if known.
    print('Applying One-Hot Encoding to categorical columns (pd.get_dummies)')
    df_encoded = pd.get_dummies(df_encoded, columns=cat_cols, drop_first=True)
else:
    print('No categorical columns to encode')

# 9. Feature Scaling
print('\n--- Feature Scaling ---')
# StandardScaler for algorithms that use distance/gradient: Logistic, KNN, SVM, KMeans, NN
scaler = StandardScaler()
X_enc = df_encoded.drop(columns=['target'])
y = df_encoded['target']
X_scaled = pd.DataFrame(scaler.fit_transform(X_enc), columns=X_enc.columns)
print('Scaled features shape:', X_scaled.shape)

# 10. Feature Selection (correlation + feature importance)
print('\n--- Feature Selection: remove highly correlated features ---')
corr_matrix = X_scaled.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
print('Dropping highly correlated features:', to_drop)
X_final = X_scaled.drop(columns=to_drop)
print('Final feature set shape:', X_final.shape)

# 11. Train-test split
print('\n--- Train-test split ---')
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, stratify=y, random_state=42)
print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)

# 12. Train Multiple Models
print('\n--- Training multiple models ---')
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000, random_state=42),
    'KNN': KNeighborsClassifier(),
    'DecisionTree': DecisionTreeClassifier(random_state=42),
    'RandomForest': RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'SVM': SVC(probability=True, random_state=42)
}
if has_xgboost:
    models['XGBoost'] = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

results = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1] if hasattr(model, 'predict_proba') else None
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc = roc_auc_score(y_test, y_proba) if y_proba is not None else None
    results.append({'model': name, 'accuracy': acc, 'precision': prec, 'recall': rec, 'f1': f1, 'roc_auc': roc})

results_df = pd.DataFrame(results).sort_values(by='f1', ascending=False).reset_index(drop=True)
print('\nModel comparison:')
print(results_df)

# Save best model (by f1)
best = results_df.loc[0, 'model']
print('\nBest model by F1:', best)
best_model = models[best]
import pickle
with open('best_model.pkl', 'wb') as f:
    pickle.dump({'model': best_model, 'scaler': scaler, 'features': X_final.columns.tolist()}, f)
print('Saved best model and scaler to best_model.pkl')

# 13. Model Evaluation (detailed for best model)
print('\n--- Detailed evaluation for best model ---')
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:,1] if hasattr(best_model, 'predict_proba') else None
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
print('\nClassification Report:')
print(classification_report(y_test, y_pred))
if y_proba is not None:
    print('\nROC AUC:', roc_auc_score(y_test, y_proba))

print('\nAll done for steps 5–14. Save files: best_model.pkl, heart_smote.csv/heart_oversampled.csv/heart_undersampled.csv (if generated)')
```
